```python
import pandas as pd
import numpy as py
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv (r"C:/Computer/ML/DataSet/archive/Google-Playstore.csv")
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df.isna().sum()
```

```python
df2 = data.dropna()
df2
```

```python
plt.figure(figsize=(30,16))
plt.xticks(rotation=90, fontsize=25)
plt.yticks(fontsize= 25)

plt.scatter (df2 ["Category"], df2 ["Installs"])
plt.xlabel ("Category")
plt.ylabel ("Installs")

plt.xlabel ("Category", fontname="Gabriola", fontsize=50)
plt.ylabel ("Installs", fontname="Gabriola", fontsize = 50)
plt.title ("Google Play Store Apps")
plt.title("Google Play Store Apps", fontname="Comic Sans MS",  fontsize = 30, color="red")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(30,16))
plt.xticks(rotation=90, fontsize=25)
plt.yticks(fontsize= 25)

plt.scatter (df2 ["Category"], df2 ["Price"])
plt.xlabel ("Category")
plt.ylabel ("Price")

plt.xlabel ("Category", fontname="Gabriola", fontsize=50)
plt.ylabel ("Price", fontname="Gabriola", fontsize = 50)
plt.title ("Google Play Store Apps")
plt.title("Google Play Store Apps", fontname="Comic Sans MS",  fontsize = 30, color="red")
plt.grid()
plt.show()
```
