```python
import  numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("C:/Computer/ML/HomeWorks/EnglandWeather.csv")
df = pd.DataFrame(data)
df.isna().sum()
dfwm = df.dropna()
dfwm.describe()
#dfwm.describe(include="all")
```

```python
dfwm.head()
```

```python
new = dfwm ["Formatted Date"].str.split(" ", n=2, expand = True)
dfwm ["Date"] = new[0]
dfwm ["Time"] = new[1]
dfwm ["NI"] = new[2]



dfwm.head()
```

```python
#dfwm.style.format({"DOB": lambda t: t.strftime("%m/%d/%Y")})
date_str = pd.to_datetime( pd.Series(dfwm["Date"]))
#dfwm ["Timestamp"] = date_str.dt.strftime ("%m/%d/%y")
dfwm ["Date"] = date_str.dt.strftime ("%m/%d/%y")

dfwm.head()
```

```python
new = dfwm ["Time"].str.split( ".", n=1, expand = True)
dfwm ["T"] = new [0]
dfwm.head()
```

```python
dfwm.pop ("T")
dfwm.pop ("NI")
dfwm.pop ("Formatted Date")
```

```python
dfwm["Year"] = pd.DatetimeIndex(dfwm['Date']).year
dfwm["Month"] = pd.DatetimeIndex(dfwm['Date']).month
dfwm["Day"] = pd.DatetimeIndex(dfwm['Date']).day

dfwm.head()
```

```python
new = dfwm ["Time"].str.split( ".", n=1, expand = True)
dfwm ["Time"] = new [0]
dfwm.head()
```

```python
plt.figure(figsize = (18,9), dpi = 80)
plt.xticks( rotation = 90, fontsize=10)
plt.yticks(fontsize= 10)
plt.yticks(range (0,120,5))
plt.xticks(range(int(np.floor(min(df['Temperature (C)']))), int(np.ceil(max(df['Temperature (C)']))),2))

plt.scatter ( dfwm ["Temperature (C)"], dfwm ["Humidity"]*100, alpha = 0.2)
plt.xlabel ("Temperature (C)", fontname="Gabriola", fontsize=20)
plt.ylabel ("Humidity", fontname="Gabriola", fontsize=20)
plt.title (" England Weather", fontname="Gabriola", color = "red", fontsize=25)
plt.grid()
plt.show()
```
